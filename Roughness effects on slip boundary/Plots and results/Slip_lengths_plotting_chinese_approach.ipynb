### Functions and correction functions

```python
def Hydro_forces(v,d,r,n):
    return (v*6*math.pi*r**2*n)*10**12/(d*10**-9)

#Correction function in the limit k --> -1. The case of a hydrophobic sphere interacting with hydrophilic sphere, b2=b and b1=0.
def correction_function_1(d,b):
    return (1/4)*(1+(3*d/(2*b))*((1+d/(4*b))*(np.log(1+(4*b)/d))-1))

#Correction function in the limit k --> 0. The case of a hydrophobic sphere interacting with similiar one, b2=b and b1=b, therefore b1=b2.
def correction_function_2(d,b):
    return (d/(3*b))*((1+d/(6*b))*(np.log(1+(6*b)/d))-1)

#Correction function in the limit k --> infinity. The case of a hydrophobic sphere interacting with a bubble, b2=b and b1=infinity, therefore b1>>b2.
def correction_function_3(d,b):
    return (d/(6*b))*((1+d/(3*b))*(np.log(1+(3*b)/d))-1)

#Corrected function
def Hydro_corrected_1(v,d,r,n,b):
    return correction_function_1(d,b)*(v*6*math.pi*r**2*n)*10**12/(d*10**-9)

def Hydro_corrected_2(v,d,r,n,b):
    return correction_function_2(d,b)*(v*6*math.pi*r**2*n)*10**12/(d*10**-9)

def Hydro_corrected_3(v,d,r,n,b):
    return correction_function_3(d,b)*(v*6*math.pi*r**2*n)*10**12/(d*10**-9)


#Correction function by chinese paper approach:
def Hydro_force_chinese_paper(v,d,r,n,b):
    return (v*6*math.pi*r**2*n)*10**12/(d*10**-9+b)
```
