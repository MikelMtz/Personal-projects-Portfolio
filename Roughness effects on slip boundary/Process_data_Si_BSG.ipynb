```python
#Shifting
Shifted_forces = []
for forces in [Force_1800nm_1um, Force_1800nm_5um, Force_1800nm_7um, Force_1800nm_10um, Force_1800nm_15um, Force_1800nm_20um, Force_1800nm_500nm, Force_1800nm_800nm]:
    forces = np.array(forces)
    min = forces.min()
    Shifted_Force = []
    for i in range(len(forces)):
        X = float(forces[i]) - min
        Shifted_Force.append(X)
    Shifted_forces.append(Shifted_Force)
```

```python
#Distance of the plate and sphere

w = 1.568*10**-9          #roughness in m (200nm)
d_plate = 0.8*10**-9    #in m
d_sphere = 2.2*10**-8  #in m

if d_sphere > 10**6*d_plate:
    d_00 = d_sphere + w

else: d_00 = d_sphere + d_plate
print(d_00)
d_0 = d_00*10**9

d_0_2 = (d_sphere + w)*10**9
print(d_0_2)
print(d_0)      #d_0 is the final distance.
```
