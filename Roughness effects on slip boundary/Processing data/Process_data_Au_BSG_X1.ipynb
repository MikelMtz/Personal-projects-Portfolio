```python
#Shifting
Shifted_forces = []
for forces in [Force_1800nm_1um, Force_1800nm_3um, Force_1800nm_5um, Force_1800nm_7um, Force_1800nm_10um, Force_1800nm_15um, Force_1800nm_20um, Force_1800nm_500nm, Force_1800nm_800nm]:
    forces = np.array(forces)
    min = forces.min()
    Shifted_Force = []
    for i in range(len(forces)):
        X = float(forces[i]) - min
        Shifted_Force.append(X)
    Shifted_forces.append(Shifted_Force)
```
